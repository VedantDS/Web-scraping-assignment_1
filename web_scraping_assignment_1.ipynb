```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

### Question 1

```python
def get_headers(url):
    
    page = requests.get(url)
    scrap = BeautifulSoup(page.content, 'html.parser')
    header = scrap.find_all('h2')
    
    headers = []
    for i in header:
        for j in i.find_all('span','mw-headline'):
            headers.append(j.get_text())
    return headers   
```

```python
get_headers('https://en.wikipedia.org/wiki/Main_Page')
```

### Question 2

```python
def get_movies1(url):
    
    page = requests.get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    name = bs.find_all('h3','lister-item-header')
    year =  bs.find_all('h3','lister-item-header')
    rating = bs.find_all('div', class_='ipl-rating-star small')
    
    ratings = []
    for i in rating:
        for j in i.find_all('span','ipl-rating-star__rating'):
            ratings.append(j.get_text())
            
    years = []
    for i in year:
        for j in i.find_all('span','lister-item-year text-muted unbold'):
            years.append(j.get_text()[1:-1])
            
    names = []
    for i in name:
        for j in i.find_all('a'):
            names.append(j.get_text())
            
    movies = pd.DataFrame({})
    movies['Name'] = names
    movies['Rating'] = ratings
    movies['Year'] = years
    movies.to_csv('top100_movies.csv',index = False)
    
    return movies
```

```python
get_movies1('https://www.imdb.com/list/ls091520106/')
```

### Question 4

```python
def get_books(url):
    
    page = requests.get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    name = bs.find_all('h4','italic')
    author = bs.find_all('div', 'flex-article-content')
    genre = bs.find_all('div', 'flex-article-content')
    
    genres = []
    for i in genre:
        for j in i.find_all('p','genre-links hidden-phone'):
            genres.append(j.get_text().replace('\n',''))

    authors = []
    for i in author:
        for j in i.find_all('p','sans bold'):
            authors.append(j.get_text()[12:-2])

    
    names = []
    for i in name:
        for j in i.find_all('a'):
            names.append(j.get_text())
            
    books = pd.DataFrame({})
    books['Name'] = names[:5]
    books['Author'] = authors[:5]
    books['Genre']  = genres[:5]
 
    return books
```

```python
get_books('https://bookpage.com/features')
```

### Question 5 (i)

```python
def get_odi_team(url):
    
    page = requests.get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    team = bs.find_all('div','rankings-block__container full rankings-table')

    teams = []
    for i in team:
        for j in i.find_all('td'):
            teams.append(j.get_text())
    teams
    teams = [tuple(teams[i:i+5]) for i in range(0, len(teams), 5)]
    odi = pd.DataFrame(teams, columns = ['Position','Team','Matches','Points','Ratings'])
    return odi
```

```python
get_odi_team('https://www.icc-cricket.com/rankings/mens/team-rankings/test')
```

### Question 7

```python
def get_phone(url):
    page = requests.get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    phone = bs.find_all('h2','a-size-mini a-spacing-none a-color-base s-line-clamp-2')
    price = bs.find_all('div','a-section a-spacing-none a-spacing-top-small') 

    phones = []
    prices = []


    for i in phone:
        for j in i.find_all('a','a-link-normal a-text-normal'):
                phones.append(i.get_text()[2:-2])

    for i in price:            
        for j in i.find_all('span','a-price-whole'):
                prices.append(j.get_text())

    product = pd.DataFrame({})
    product['Product']  = phones
    product['prices'] = prices
    return product
```

```python
get_phone('https://www.amazon.in/s?k=smartphone+under+20000&ref=nb_sb_noss_1')
```
